```python
# %load C:\Users\Walter\Desktop\tools\template\plot_template.py

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

plt.rc('font', family='serif')
plt.rc('axes', labelsize=16)
plt.rc('xtick', labelsize=14, color='grey')
plt.rc('ytick', labelsize=14, color='grey')
plt.rc('legend', fontsize=16, loc='lower left')
plt.rc('savefig', dpi=330, bbox='tight')
%matplotlib inline
```

```python
CT_status = pd.read_csv('data/CT_status.csv', index_col=0).drop('Excel Time',axis=1)
CT_status.index = pd.to_datetime(CT_status.index)
```

```python
CT_status = CT_status.resample('10s').max().fillna(method='pad')
```

```python
CT_status.mean()
```

```python
start_date='2018-09-10'
end_date='2018-09-17'
CT_status.resample('30s').max().fillna(method='pad').truncate(before=start_date,after=end_date).plot()
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
