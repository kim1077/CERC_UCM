```python
from pandas import *
```

```python
# 1 hr sampling time data
dd=date_range('2018', '2018-01-01 02:00',freq='60T')
print(dd)
```

```python
d0=DataFrame(data={'a':[1,2,3],'b':[10,20,30]})
d0=d0.set_index(dd)
d0
```

```python
# interpolation & extrapolation 
dn=date_range('2018', '2018-01-01 03:00',freq='15T')
dhigh=d0.reindex(dn,method='nearest')
dhigh
```

```python
d=date_range('2018', '2018-01-01 03:00', freq='1H')
dlow=dhigh.reindex(d)
dlow
help(dhigh.reindex)
```
